```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.random.random(50) * 100
Y = np.random.random(50) * 100
```

# Scatter plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(x=X, y=Y, c='g', marker='*', s=100, alpha=0.5, label='X vs, Y')
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.title("X vs, Y", fontsize=20)
plt.legend(loc='upper left')
plt.show()
```

# Histograms

```python
ages =  np.random.normal(20, 1.5, 1000)
```

```python
plt.hist(ages, bins=30)
plt.show()
```

```python
plt.hist(ages, bins=[ages.min(), 18, 21, ages.max()])
plt.show()
```

```python
plt.hist(ages, bins=20, cumulative=True)
plt.show() 
```

# Pie Charts

```python
langs = ['Python', 'C++','Java', 'C#', "Go"]
votes = [50, 24, 14, 6, 17]
```

```python
plt.pie(votes, labels=langs)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(votes, labels=None)
plt.legend(labels=langs)
plt.show()
```

```python
explodes = [0.3, 0,0,0, 0]
```

```python
plt.pie(votes, labels=langs, explode=explodes)
plt.show()
```

```python
explodes = [0, 0, 0, 0.3, 0]
```

```python
plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%', pctdistance=0.8)
plt.show()
```

```python
plt.pie(votes, labels=langs, startangle=90, autopct='%.2f%%', pctdistance=0.8)
plt.show()
```

# multiple figure 

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)
```

```python
plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2,2,figsize=(10, 6))

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sine Wave')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cosine Wave')

axs[1, 0].plot(x, np.random.random(100) )
axs[1, 0].set_title('Random Wave')

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title('log function')

fig.suptitle('Four plots')
               
plt.show()
```

# 3D Ploting

```python
ax = plt.axes(projection='3d')
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title('3d Plot')
ax.set_xlabel('test')
plt.show()
```

```python
ax = plt.axes(projection='3d')
x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set_title('3d Plot')
ax.set_xlabel('test')
plt.show()
```
